```python
import pandas as pd
X_test_df = pd.read_csv(r'C:\Users\Sergio Orozco\Desktop\OReillyBookCodeAlong\Chapter 3 - Classification\Exercise 1\testing_data.csv')
y_test_df = pd.read_csv(r'C:\Users\Sergio Orozco\Desktop\OReillyBookCodeAlong\Chapter 3 - Classification\Exercise 1\testing_labels.csv')
X_test = X_test_df.drop(X_test_df.columns[0], axis=1).to_numpy()
y_test = y_test_df.drop(y_test_df.columns[0], axis=1).to_numpy().ravel()
```

```python
from sklearn.externals import joblib

knn_best_model = joblib.load(r"C:\Users\Sergio Orozco\Desktop\OReillyBookCodeAlong\Chapter 3 - Classification\Exercise 1\knn_best_model.pkl")
```

```python
y_predictions = knn_best_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
conf_mat = confusion_matrix(y_test,y_predictions)
plt.matshow(conf_mat, cmap=plt.cm.gray)
plt.show()
```

```python
from sklearn.metrics import f1_score

f1_score(y_test,y_predictions,average="macro")
```

```python
row_sums = conf_mat.sum(axis=1, keepdims=1)
norm_conf_mat = conf_mat / row_sums
```

```python
import numpy as np
np.fill_diagonal(norm_conf_mat,0)
plt.matshow(norm_conf_mat, cmap=plt.cm.gray)
plt.show()
```
